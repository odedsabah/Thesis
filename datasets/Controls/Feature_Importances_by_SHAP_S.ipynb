```python
# Upload packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import shap 

from sklearn import metrics
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
pcoa_all2rf = pd.read_csv("~/metaanalysis/datasets/Controls/shap_rf_s.csv")
cluster_manually = pd.read_csv("~/metaanalysis/datasets/Controls/cluster_manually.csv")
```

```python
sns.scatterplot(data=cluster_manually, x='x', y='y', hue='cluster')

plt.show() 
```

```python
pcoa_all2rf['cluster'].value_counts()
```

```python
filtered_columns = [col for col in pcoa_all2rf.columns if col.startswith('s__') and not col.startswith('s__GG')] + ['cluster']
pcoa_all2rf = pcoa_all2rf[filtered_columns]
```

```python
np.random.seed(888)

pcoa_all2rf['cluster'] = pd.Categorical(pcoa_all2rf['cluster'])
pcoa_all2rf['cluster_encoded'] = pcoa_all2rf['cluster'].cat.codes
pcoa_all2rf['cluster_encoded'] = 'Level_' + (pcoa_all2rf['cluster_encoded'] + 1).astype(str)

X = pcoa_all2rf.drop(['cluster', 'cluster_encoded'], axis=1)  
y = pcoa_all2rf['cluster_encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=888)

RF_model = RandomForestClassifier(max_depth=None,
                                  min_samples_split=5, random_state=888)

n_estimators_RF = [100, 200, 300, 400, 500]
max_features_RF = [2,4,8,16,32,None]
parameters_RF = {'n_estimators': n_estimators_RF,
                'max_features': max_features_RF}

cv_rf = GridSearchCV(estimator=RF_model, param_grid=parameters_RF, cv=5, scoring='accuracy')
cv_rf.fit(X_train, y_train)

best_rf_model = cv_rf.best_estimator_

predictions = best_rf_model.predict(X_test)

conf_matrix = confusion_matrix(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)

print("Confusion Matrix:")
print(conf_matrix)

print("\nAccuracy:", accuracy)
```

```python
explainer = shap.TreeExplainer(best_rf_model)
shap_values = explainer.shap_values(X_test, check_additivity=False)
```

```python
shap_interaction_values = explainer.shap_interaction_values(X_test)
```

```python
for cluster in range(4):
    interaction_values_sum = np.sum(shap_interaction_values[cluster], axis=2)
    np.fill_diagonal(interaction_values_sum, 0) 

    shap.summary_plot(interaction_values_sum, X_test)
```

```python
for cluster in range(4):

    shap_values_for_class = shap_interaction_values[cluster]
    tmp = shap_values_for_class.sum(axis=0)
    for i in range(tmp.shape[0]):
        tmp[i, i] = 0

    N = 5 
    top_feature_indices = np.argsort(-tmp.sum(axis=0))[:N]

    tmp2 = tmp[top_feature_indices, :][:, top_feature_indices]

    plt.figure(figsize=(12, 12))
    plt.imshow(tmp2, cmap='OrRd')
    plt.colorbar()
    plt.yticks(range(N), [X.columns[i] for i in top_feature_indices], rotation=0, horizontalalignment="right")
    plt.xticks(range(N), [X.columns[i] for i in top_feature_indices], rotation=90, horizontalalignment="left")
    plt.gca().xaxis.tick_top()
    plt.show()
```

```python
"""This program get metagenomic data on MetaPhlan4 format after dropping the feature, it calculates the top of feature importance to evaluate the accuracy.
When the accuracy is less than 90%, break the condition."""

METAPHLAN4MGX = pcoa_all2rf
important_species = {}
Accuracy = 100
while Accuracy > 90:

    MGX_X = METAPHLAN4MGX.drop(['cluster', 'cluster_encoded'], axis=1)  
    MGX_y = METAPHLAN4MGX['cluster_encoded']

    X_train, X_test, y_train, y_test = train_test_split(MGX_X, MGX_y, test_size=0.30, random_state=888)
    
    n_estimators_RF = [200]
    max_features_RF = [None]
    parameters_RF = {'n_estimators': n_estimators_RF,
                    'max_features': max_features_RF}
    RF_model = RandomForestClassifier(max_depth=None,
                                      min_samples_split=5, random_state=888)
    RF_clf = GridSearchCV(RF_model, parameters_RF, cv=5, scoring= 'accuracy')
    RF_clf.fit(X_train, y_train)
    print("Best Model:")
    RF_model = RF_clf.best_estimator_
    print(RF_model)
    RF_predictions = RF_model.predict(X_test)
    
##########################################################################

    imp_df = pd.DataFrame({
    "": MGX_X.columns,
    "feature_importances": RF_clf.best_estimator_.feature_importances_
    })
    feature_importances = imp_df.sort_values(by="feature_importances", ascending=False).rename(columns={"":"feature_name"})

##########################################################################

    dominant_species = dict(zip(feature_importances.feature_name.iloc[:1], feature_importances.feature_importances.iloc[:1]))
    METAPHLAN4MGX = METAPHLAN4MGX.drop(list(dominant_species.keys()), axis=1) 
    Accuracy = round(metrics.accuracy_score(y_test, RF_predictions) * 100, 2)
    important_species[frozenset(dominant_species.keys())] = f'{Accuracy}%'
    print(f'Number of columns after drop:{len(MGX_X.columns)}')
    print(dominant_species)
    print(f'Accuracy {Accuracy}%')
    print(feature_importances.head())

    print("########################################################################")
important_species = pd.DataFrame.from_dict(important_species, orient='index', columns=["Accuracy"])
print(important_species)
```
